# First Moment of Business

```python
col_for_analysis = ['Open','High','Low','Close','TrendSupertrend(7,3)','Volume']
col_for_analysis_data = data[col_for_analysis]

#mean
mean = col_for_analysis_data.mean()
print("Mean: \n",mean)
print("-----------------------------------------------------------------------")

#median
median = col_for_analysis_data.median()
print("Median: \n",median)
print("-----------------------------------------------------------------------")

#mode
mode = col_for_analysis_data.mode()
print("Mode: \n",mode)
```

# Second Moment of Business

```python
#variance
variance = col_for_analysis_data.var()
print("Variance: \n",variance)
print("-----------------------------------------------------------------------")

#standard deviation
std_deviation = col_for_analysis_data.std()
print("Standard Deviation: \n",std_deviation)
print("-----------------------------------------------------------------------")

#range
col_range = col_for_analysis_data.max() - col_for_analysis_data.min()
print("Range: \n",col_range)
```

# Third Moment of Business Decision

```python
#Skewness
skew_data = col_for_analysis_data.skew()
print("Skewness: \n",skew_data)
```

# Fourth Moment of Business

```python
#kurtosis
Kurtosis_data = col_for_analysis_data.kurt()
print("Kurtosis: \n",Kurtosis_data)
```

```python
#outlier treatement on volume

Volume_analysis = data['Volume']
Q1 = Volume_analysis.quantile(0.25)
Q3 = Volume_analysis.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - (IQR * 1.5)
upper_limit = Q3 + (IQR * 1.5)

outlier = Volume_analysis[(Volume_analysis < lower_limit) | (Volume_analysis > upper_limit)]
outlier
```

```python
#trimming of outliers
lower_value = Volume_analysis.quantile(0.05)
upper_value = Volume_analysis.quantile(0.95)
outliers = Volume_analysis[(Volume_analysis < lower_value) | (Volume_analysis > upper_value)]
```

```python
trimmed_data = Volume_analysis[(Volume_analysis >= lower_value) & (Volume_analysis <= upper_value)]
```

```python
data.shape, trimmed_data.shape
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(trimmed_data, vert=False)
plt.xlabel('Volume')
plt.title('Volume after Trimming Outliers')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(trimmed_data, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Trimmed Volume Data')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```
